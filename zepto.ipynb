```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
df.head(10)
```

```python
df.drop(columns=["total_revenue","discount"], inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"current_price": "selling_price","orders":"order_count"}, inplace=True)
```

```python
df.head()
```

```python
df["total_revenue"] = df["selling_price"] * df["order_count"]
```

```python
df["total_profit"] = df["total_revenue"] - (df["original_price"] * df["order_count"])
```

```python
df["profit_per_order"] = df["total_profit"] / df["order_count"]
```

```python
df.head()
```

```python
df["influencer_active"] = df["influencer_active"].map({"Yes": 1, "No": 0})
```

```python
df = df.drop_duplicates()
```
